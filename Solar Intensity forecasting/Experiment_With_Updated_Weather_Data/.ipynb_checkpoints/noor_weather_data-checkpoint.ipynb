```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("noor_weather_data.csv")
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
data_num = data.select_dtypes(include=['datetime', np.number]).dropna(
    axis=1, how='all').interpolate().ffill().bfill().set_index('datim')
date_range_10_min = pd.date_range(start=data['datim'].min(), SSend=data['datim'].max(), freq='10T')
```

```python
# final_df = pd.DataFrame(index=date_range_10_min).join(data_num, how='outer').ffill(limit=9).bfill(limit=9)
final_df = pd.DataFrame(index=date_range_10_min).join(data_num, how='outer').ffill(limit=5).bfill(limit=5).interpolate()
final_df = final_df.reset_index().rename(columns={'index': 'datim'})
```

```python
final_df.to_csv('D:/Projects/weather_data/noor_weather_data_10min_numeric_only.csv', index=False)
```
